# Constructing model and predicting ratings

```python
# Prediction
def predict_rating(ratings_train, ratings_valid):
    # Define trainset and testset
    X_train = ratings_train.copy()
    X_test = ratings_valid.copy()

    # Define a reader
    reader = Reader(rating_scale=(0, 5))

    # Load data into a Surprise dataset
    data = Dataset.load_from_df(X_train[['userId', 'movieId', 'rating']], reader)

    # Initialize SVD algorithm
    algo = SVD()

    # Fit the algorithm on the training data
    trainset = data.build_full_trainset()
    algo.fit(trainset) #fill the ratings with SVD method

    # Predictions on the test set
    testset = [(row['userId'], row['movieId'], row['rating']) for index, row in X_test.iterrows()]
    predictions = algo.test(testset)

    # Extract predicted ratings
    predicted_ratings = [pred.est for pred in predictions]

    return predicted_ratings
```

```python
# Predict ratings
predicted_ratings = predict_rating(ratings_train, ratings_valid)
print("Predicted ratings:", predicted_ratings)
```

```python
from sklearn.metrics import mean_squared_error

r_true = ratings_valid['rating'].to_numpy()
r_pred = predicted_ratings

rmse = mean_squared_error(r_true, r_pred, squared=False)
print(f"RMSE = {rmse:.4f}")
```
